### <font color =blue>Task 2: Instructions 

Clean the dataset.

 - Create a list named chars_to_remove that contains the following characters: + , and $.
 - Create a list named cols_to_clean that contains the following column names: Installs and Price.
 - For each column in cols_to_clean in the apps DataFrame, replace each character in chars_to_remove with the empty string ''. 

Note: Make sure to use an empty string '' and not a space character ' '
 - Finally, print a summary of the apps dataframe using the info() function. Observe the output. 

Note: Notice that Installs and Price are still of type object and not int or float as we would have expected after removal of the special characters. We will solve this issue in the next task.

Helpful links:

 - pandas apply() documentation
 - Python replace() documentation

```python
chars_to_remove = ['+', ',', '$']
cols_to_clean = ['Installs', 'Price']
for columns in cols_to_clean: 
    for chars in chars_to_remove:
        apps[columns] = apps[columns].apply(lambda x:x.replace(chars, '')) 
#The apply() method is used to apply the function 'lambda' to each value in the columns specified in the df.   
#The lambda function returns the result of the replace() method applied to x, which replaces the characters specified by the inner for loop with an empty string ''.
#x is a parameter name representing the individual value in the columns specified by the outer for-loop that is being processed by the apply() method
    print(apps.info())
```

#### <font color = blue> Task 6: Instructions
Examine the relationship between size, price, and rating of apps using jointplot(). 

Recall from Task #1 that we had observed some missing values in the Rating and Size columns. To make rational decisions, it is important that we do not consider these missing values in our analysis. We will work with a subset apps_with_size_and_rating_present DataFrame for this task.

 - Select rows from apps where both Rating and Size values are present, ie - they are not null. Store the result in the apps_with_size_and_rating_present dataframe.
 - From apps_with_size_and_rating_present, select the categories having atleast 250 apps. Assign the result to large_categories dataframe.
 - Fill out x and y to create a joint plot of Rating as a function of Size.
 - From apps_with_size_and_rating_present dataframe, select all Paid apps. Save the result in paid_apps.
 - Fill out x and y to create a joint plot of Rating as a function of Price. 

Helpful links:

 - There are many ways to subset a dataframe and select rows based on a column value. This exercise from Data Manipulation with pandas may be a good starting place.
 - jointplot() documentation

```python
apps.info()
```

```python
#selecting the rows in rating and size column that are not NaN

apps_with_size_and_rating_present = apps[(apps['Rating'].isna() == False) & (apps['Size'].isna() == False)]
apps_with_size_and_rating_present.info()
```

```python
#here I filter out the rows that contain 'Varied with device' in the Size column so I can perform arithmetic operations on the rest

apps_with_size_and_rating_present = apps_with_size_and_rating_present[apps_with_size_and_rating_present['Size'] != 'Varies with device']
apps_with_size_and_rating_present
```

```python
#the size column has the M = MB and k = kb shwoing size, 1000kb make 1MB. So we replace the characters with an empty string and 
#convert all to one unit (MB) by dividing the rows containing k by 1000

apps_with_size_and_rating_present['Size'] = np.where(apps_with_size_and_rating_present['Size'].str.contains('M'), apps_with_size_and_rating_present['Size'].str.replace('M', ''), apps_with_size_and_rating_present['Size'])

apps_with_size_and_rating_present['Size'] = np.where(apps_with_size_and_rating_present['Size'].str.contains('k'), apps_with_size_and_rating_present['Size'].str.replace('k', '').astype('float')/1000, apps_with_size_and_rating_present['Size'])

```

```python
apps_with_size_and_rating_present['Size'].unique()  #to confirm the result of our 2 previous steps
```

```python
#convert size column data  type to float

apps_with_size_and_rating_present['Size'] = apps_with_size_and_rating_present['Size'].astype('float')
apps_with_size_and_rating_present.dtypes
```

```python
#line1 groups the apps_with_size... df by its category column, getting each group's size. each size is represented by the nos or rows they appear in, into the df apps_category
#line2 creates a df that filters the apps_category df for those category groups whose size is 250 upwards
#line3- the apps_with_size... df is filtered against the filtered_cat... df so that where values in the'Category' column is one of the categories that have at least 250 apps, the result is saved in the new df 'Large_categories'

apps_category = apps_with_size_and_rating_present.groupby('Category').size()
filtered_category = apps_category[apps_category.values >= 250]
large_categories = apps_with_size_and_rating_present[apps_with_size_and_rating_present['Category'].isin(list(filtered_category.index))]
large_categories
```

```python
sns.jointplot(x = large_categories['Size'], y = large_categories['Rating'])
plt.title('Plot showing Rating as a function of Size', loc = 'right', y = 1.2)
```

```python
#selection of all paid apps
paid_apps = apps_with_size_and_rating_present[apps_with_size_and_rating_present['Type'] == 'Paid']
paid_apps
```

```python
sns.jointplot(x = paid_apps['Rating'], y = paid_apps['Price'])
plt.title('Plot of Rating as a function of Price', loc = 'right', y = 1.2)
```

#### <font color = blue> Task 9: Instructions
Prep the data for a box plot that compares the number of installs of paid apps vs. number of installs of free apps.

 - From apps, filter rows where for Type == Paid, and select the Installs column and assign it to y of trace0.
 - From apps, filter rows where for Type == Free, and select the Installs column and assign it to y of trace1. 

Helpful links:

 - Plotly box plot documentation
 - Interpreting box plots article

```python
apps_type = apps.copy()

apps_paid = apps_type[apps_type['Type']== 'Paid']
apps_paid.loc[:,'Installs_log'] = np.log(apps_paid.loc[:,'Installs'])

free_apps = apps_type[apps_type['Type']== 'Free']
free_apps.loc[:,'Installs_log'] = np.log(free_apps.loc[:,'Installs'])
                                             
import plotly.graph_objs as go

trace0 = go.Box(x = apps_paid['Type'], y = apps_paid['Installs_log'], name = 'Paid')
trace1 = go.Box(x = free_apps['Type'], y = free_apps['Installs_log'], name = 'Free')

layout = go.Layout(title = 'Number of Installs for Paid vs Free Apps',
                  xaxis_title = 'App Type',
                  yaxis_title = 'Number of Installs (Log Scale)')
data = [trace0, trace1]
fig = go.Figure(data= data, layout = layout)
fig.show()
```

#### <font color = blue>Popularity of paid apps vs free apps¶ 

For apps in the Play Store today, there are five types of pricing strategies: free, freemium, paid, paymium, and subscription. Let's focus on free and paid apps only. Some characteristics of free apps are: 
  - Free to download. 
  - Main source of income often comes from advertisements. 
  - Often created by companies that have other products and the app serves as an extension of those products. 
  - Can serve as a tool for customer retention, communication, and customer service. 

Some characteristics of paid apps are: 
  - Users are asked to pay once for the app to download and use it. 
  - The user can't really get a feel for the app before buying it. 

Are paid apps installed as much as free apps? It turns out that paid apps have a relatively lower number of installs than free apps, though the difference is not as stark as I would have expected!

User reviews for apps can be analyzed to identify if the mood is positive, negative or neutral about that app. For example, positive words in an app review might include words such as 'amazing', 'friendly', 'good', 'great', and 'love'. Negative words might be words like 'malware', 'hate', 'problem', 'refund', and 'incompetent'.
